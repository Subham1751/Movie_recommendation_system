```python
import pandas as pd
import numpy as np
import difflib
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

**Data collection and Pre-processing**

```python
movies_data = pd.read_csv("movies.csv")
movies_data.head(2)
```

```python
movies_data.shape
```

```python
movies_data.columns
```

```python
movies_data.isnull().sum()
```

**Feature Selcetion**

```python
relevant_features = ['genres' , 'keywords', 'tagline' , 'cast' , 'director' , 'overview', 'title']
print(relevant_features)
```

```python
# Fill null values in relevent features list with empty string.

for feature in relevant_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combine all the selected features.

combined_features = movies_data['title']+ ' ' + movies_data['genres']+ ' ' + movies_data['overview']+ ' ' + movies_data['keywords']+ ' ' + movies_data['tagline']+ ' ' + movies_data['cast']+ ' ' + movies_data['director']
```

```python
print(combined_features)
```

**Convert Text data to numerical vector**

```python
# using TfidVectorizer we can Converts text into vectors

vectorizer = TfidfVectorizer()

feature_vector = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

**Cosine Similarity**

```python
# Getting the similarity scores using cosine_similarity : Measures similarity between two vectors using the cosine of the angle between them.

similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
# This is the similarity score of each movie comparing with other movies
```

```python
# we can check the shape also for this

print(similarity.shape)
```

**Getting name of movie from user**

```python
movie_name = input('Search For Movie = ')
```

```python
# Creating a list with all the movies name given in dataset

list_of_all_movies = movies_data['title'].tolist()
print(list_of_all_movies)
```

```python
# Finding the closest match of the movie entered by user
```

**Using difflib**

```python
# use difflib.get_close_match and pass the parameters as movie entered by user and second as list of all movies

find_closest_match = difflib.get_close_matches(movie_name , list_of_all_movies)
print(find_closest_match)
```

```python
# selecting only the first movie matched with the movie entered by user

top_match = find_closest_match[0]
print(top_match)
```

```python
# Finding the index value of movie with title

index_of_movie = movies_data[movies_data.title == top_match]['index'].values[0]
print(index_of_movie)
```

**Using similarity to find the similarity score**

```python
# Getting the similarity score for the list of movies similar to the movie entered by user

similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

**Sorting the movie based on similarity score (descending order)**

```python
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True)
print(sorted_similar_movies)
```

```python
# Print the name of suggested or similar movies based on the index value 
# we will create a for loop to print the first 10 recommended movies 
```

```python
print('Recommended Movies for you : \n')

i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]

    if i<11:
        print(i,'.', title_from_index)
        i+=1
```

# Movie Recommendation System

```python
movie_name = input('Search For Movie = ')

find_closest_match = difflib.get_close_matches(movie_name, list_of_all_movies, n=3, cutoff=0.5)

if find_closest_match:   # Clean check first
    try:
        top_match = find_closest_match[0]
        index_of_movie = movies_data[movies_data.title == top_match]['index'].values[0]

        similarity_score = list(enumerate(similarity[index_of_movie]))
        sorted_similar_movies = sorted(similarity_score, key=lambda x: x[1], reverse=True)

        print("Recommended movies for you : \n")
        i = 1
        for movie in sorted_similar_movies:
            index = movie[0]
            title_from_index = movies_data[movies_data.index == index]['title'].values[0]

            if i < 16:
                print(i, '.', title_from_index)
                i += 1

    except Exception as e:   # catches any unexpected errors
        print("An error occurred while generating recommendations:", e)

else:
    print("Movie not found in dataset.")
    print("Did you mean one of these?")
    print(find_closest_match if find_closest_match else "No close matches found.")
```
